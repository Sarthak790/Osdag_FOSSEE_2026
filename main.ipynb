```python
# Extracting the central elements and ordering them based on connectivity

central_elements = [15,24,33,42,51,60,69,78,83]

ordered_elements = []
remaining = central_elements.copy()

current = remaining.pop(0)
ordered_elements.append(current)

while remaining:
    last_node = members[current][1]
    for el in remaining:
        if members[el][0] == last_node:
            ordered_elements.append(el)
            current = el
            remaining.remove(el)
            break

print("Ordered central elements:", ordered_elements)

```

```python
# Extracting the forces for the ordered central elements
Mz_i = []
Mz_j = []
Vy_i = []
Vy_j = []

for el in ordered_elements:
    Mz_i.append(float(df["forces"].sel(Element=el, Component="Mz_i")))
    Mz_j.append(float(df["forces"].sel(Element=el, Component="Mz_j")))
    Vy_i.append(float(df["forces"].sel(Element=el, Component="Vy_i")))
    Vy_j.append(float(df["forces"].sel(Element=el, Component="Vy_j")))

print("Mz_i:", Mz_i)
print("Mz_j:", Mz_j)
print("Vy_i:", Vy_i)
print("Vy_j:", Vy_j)
```

```python
# Constructing the node chain along the central elements
node_chain = [members[ordered_elements[0]][0]]

for el in ordered_elements:
    node_chain.append(members[el][1])

print("Node chain along central elements:", node_chain)
```

```python
# Extracting the x-coordinates of the nodes in the chain
x_positions = [nodes[n][0] for n in node_chain]
print("X positions of central nodes:", x_positions)
```

```python
# Plotting Mz and Vy along the central elements
Mz_plot = [Mz_i[0]]
Vy_plot = [Vy_i[0]]

for i in range(len(ordered_elements)):
    Mz_plot.append(Mz_j[i])
    Vy_plot.append(Vy_j[i])
```

```python
# Plotting the bending moment diagram (Mz) along the central elements
import plotly.graph_objects as go
import numpy as np

x = np.array(x_positions)
Mz = np.array(Mz_plot)

fig = go.Figure()

fig.add_trace(go.Scatter(
    x=x,
    y=Mz,
    mode="lines",
    line=dict(color="firebrick", width=2),
    name="BMD"
))

# Adding vertical lines for each point to enhance visibility
for xi, yi in zip(x, Mz):
    fig.add_shape(
        type="line",
        x0=xi, x1=xi,
        y0=0, y1=yi,
        line=dict(
            color="rgba(0, 70, 200, 0.9)",  # strong blue
            width=1.5                       # slightly thicker
        )
    )

# Adding a horizontal line at y=0 for reference
fig.add_hline(y=0, line=dict(color="black", width=2))


Mz_max = Mz.max()
Mz_min = Mz.min()

fig.add_annotation(
    x=x[np.argmax(Mz)],
    y=Mz_max,
    text=f"Max: {Mz_max:.2f} kN·m",
    showarrow=True,
    arrowhead=2
)

fig.add_annotation(
    x=x[np.argmin(Mz)],
    y=Mz_min,
    text=f"Min: {Mz_min:.2f} kN·m",
    showarrow=True,
    arrowhead=2
)

# Final layout adjustments
fig.update_layout(
    template="simple_white",
    title="Bending Moment Diagram",
    xaxis_title="Span (m)",
    yaxis_title="Moment (Mz)",
    height=450,
    width=1000,
    showlegend=False
)

fig.update_xaxes(showgrid=False, zeroline=False)
fig.update_yaxes(showgrid=False)

fig.show()
```

```python
# Defining the girders with their respective elements and nodes
girders = {
    "Girder 1": {"elements": [13,22,31,40,49,58,67,76,81],
                 "nodes": [1,11,16,21,26,31,36,41,46,6]},
    "Girder 2": {"elements": [14,23,32,41,50,59,68,77,82],
                 "nodes": [2,12,17,22,27,32,37,42,47,7]},
    "Girder 3": {"elements": [15,24,33,42,51,60,69,78,83],
                 "nodes": [3,13,18,23,28,33,38,43,48,8]},
    "Girder 4": {"elements": [16,25,34,43,52,61,70,79,84],
                 "nodes": [4,14,19,24,29,34,39,44,49,9]},
    "Girder 5": {"elements": [17,26,35,44,53,62,71,80,85],
                 "nodes": [5,15,20,25,30,35,40,45,50,10]},
}
```

```python
# Function to build a 3D force diagram (BMD or SFD) in a MIDAS-like style
def build_3d_force_diagram(
    df,
    nodes,
    girders,
    component="Mz",            # "Mz" for BMD, "Vy" for SFD
    scale_ribbon=0.08,         # extrusion scale for ribbon
    scale_curtain=0.30,        # extrusion scale for vertical lines
    n_strips=10,               # ribbon thickness
    n_vertical=60,             # curtain density
    cmap_name="jet"
):
   
    import numpy as np
    import plotly.graph_objects as go
    from matplotlib import cm
    from matplotlib.colors import Normalize

    cmap = cm.get_cmap(cmap_name)

    
    all_forces = []
    for data in girders.values():
        for el in data["elements"]:
            all_forces.append(float(df["forces"].sel(Element=el, Component=f"{component}_i")))
            all_forces.append(float(df["forces"].sel(Element=el, Component=f"{component}_j")))

    fmin, fmax = min(all_forces), max(all_forces)
    norm = Normalize(vmin=fmin, vmax=fmax)

    fig = go.Figure()

    # Loop through each girder and build the 3D force diagram
    for girder_name, data in girders.items():

        elements = data["elements"]
        node_chain = data["nodes"]

        
        Fi, Fj = [], []
        for el in elements:
            Fi.append(float(df["forces"].sel(Element=el, Component=f"{component}_i")))
            Fj.append(float(df["forces"].sel(Element=el, Component=f"{component}_j")))

        F = np.array([Fi[0]] + [-v for v in Fj])

        
        x = np.array([nodes[n][0] for n in node_chain])
        y = np.array([nodes[n][1] for n in node_chain])
        z = np.array([nodes[n][2] for n in node_chain])

        y_ribbon = y + scale_ribbon * F

        
        fig.add_trace(go.Scatter3d(
            x=x, y=y, z=z,
            mode="lines",
            line=dict(color="rgba(130,130,130,0.8)", width=3),
            showlegend=False
        ))

        # Adding the colored ribbon for the force diagram
        for k in range(n_strips):
            offset = (k - n_strips/2) * 0.025

            colors = [
                f"rgb{tuple((np.array(cmap(norm(val)))[:3]*255).astype(int))}"
                for val in F
            ]

            fig.add_trace(go.Scatter3d(
                x=x,
                y=y_ribbon + offset,
                z=z,
                mode="lines",
                line=dict(color=colors, width=4),
                showlegend=False
            ))

        # Adding the vertical lines (curtain) for the force diagram
        x_dense = np.linspace(x.min(), x.max(), n_vertical)
        F_dense = np.interp(x_dense, x, F)
        z_dense = np.interp(x_dense, x, z)

        for xi, zi, fi in zip(x_dense, z_dense, F_dense):
            yi = scale_curtain * fi
            color = f"rgb{tuple((np.array(cmap(norm(fi)))[:3]*255).astype(int))}"

            fig.add_trace(go.Scatter3d(
                x=[xi, xi],
                y=[0, yi],
                z=[zi, zi],
                mode="lines",
                line=dict(color=color, width=4),
                showlegend=False
            ))

    # Adding a dummy trace for the colorbar
    fig.add_trace(go.Scatter3d(
        x=[None], y=[None], z=[None],
        mode='markers',
        marker=dict(
            colorscale=cmap_name,
            cmin=fmin,
            cmax=fmax,
            colorbar=dict(
                title=f"{component} Force",
                thickness=15
            ),
            showscale=True
        ),
        showlegend=False
    ))

    # Final layout adjustments
    fig.update_layout(
        title=f"3D {'Bending Moment' if component=='Mz' else 'Shear Force'} Diagram (MIDAS-style)",
        scene=dict(
            xaxis=dict(title="X (Span)", showbackground=False),
            yaxis=dict(title="Force Magnitude", showbackground=False),
            zaxis=dict(title="Z (Transverse)", showbackground=False),
            aspectmode="data"
        ),
        height=900,
        showlegend=False
    )

    return fig
```

```python
fig_bmd = build_3d_force_diagram(
    df=df,
    nodes=nodes,
    girders=girders,
    component="Mz"
)
fig_bmd.show()
```
